```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
original_data=pd.read_csv("C:/Users/李宗泽/Desktop/amazon.csv")
original_data.head()
```

```python
cleaned_data=original_data.copy()
cleaned_data.head()
```

```python
cleaned_data.info()
```

```python
# Convert prices to float
cleaned_data['discounted_price'] = (
    cleaned_data['discounted_price'].str.replace("₹", "").str.replace(",", "").astype(float))
cleaned_data['actual_price'] = (
    cleaned_data['actual_price'].str.replace("₹", "").str.replace(",", "").astype(float))
# Convert discount percentage to fraction
cleaned_data['discount_percentage'] = (
    cleaned_data['discount_percentage'].str.replace("%", "").astype(float) / 100)
# Clean & convert ratings
cleaned_data['rating'] = cleaned_data['rating'].replace("|", "3.9").astype(float)
# Convert rating_count to numeric
cleaned_data['rating_count'] = cleaned_data['rating_count'].str.replace(",", "").astype(float)
cleaned_data.info()
```

```python
high_rating_data=cleaned_data[cleaned_data['rating']>=4.0]
sorted_high_rating=high_rating_data.sort_values(by='rating',ascending=False)
sorted_high_rating
```

```python
data_1=sorted_high_rating.copy()
data_1.head()
```

```python
data_1.drop(columns=['user_id','user_name','review_id','img_link','product_link'],axis=1,inplace=True)
```

```python
data_1.head()
```

```python
data_2=data_1.copy()
```

```python
data_2.drop(columns=['about_product','review_title','review_content'],axis=1,inplace=True)
```

```python
data_2.head()
```

```python
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 提取category的一级分类（原分类是多层级，取第一部分）
data_2['category_level1'] = data_2['category'].str.split('|').str[0]
```

```python
# 1. 品类分布：看核心业务品类
# 图表类型：水平条形图作用：识别主力品类，明确业务布局重点
plt.figure(figsize=(10, 6))
category_counts = data_2['category_level1'].value_counts()
sns.barplot(x=category_counts.values, y=category_counts.index, palette='viridis')
plt.title('各品类产品数量分布', fontsize=14)
plt.xlabel('产品数量', fontsize=12)
plt.ylabel('品类', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 2. 价格对比：看原价 vs 折扣价差额
# 图表类型：双轴条形图作用：直观识别 “高优惠力度” 产品
# 取前10个产品展示（避免拥挤）
sample_data = data_2.head(10)

fig, ax1 = plt.subplots(figsize=(12, 6))
# 原价（左轴）
ax1.bar(sample_data['product_id'], sample_data['actual_price'], color='lightcoral', label='原价')
ax1.set_xlabel('产品名称', fontsize=12)
ax1.set_ylabel('原价', fontsize=12, color='lightcoral')
ax1.tick_params(axis='y', labelcolor='lightcoral')
plt.xticks(rotation=45, ha='right')

# 折扣价（右轴）
ax2 = ax1.twinx()
ax2.bar(sample_data['product_id'], sample_data['discounted_price'], color='lightblue', alpha=0.7, label='折扣价')
ax2.set_ylabel('折扣价', fontsize=12, color='lightblue')
ax2.tick_params(axis='y', labelcolor='lightblue')

fig.legend(loc='upper right')
plt.title('产品原价vs折扣价对比（前10个产品）', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 3. 折扣率分布：看整体优惠策略
# 图表类型：直方图作用：了解多数产品的折扣力度集中区间
plt.figure(figsize=(10, 6))
sns.histplot(data_2['discount_percentage'], bins=8, kde=True, color='green')
plt.title('折扣率分布', fontsize=14)
plt.xlabel('折扣率', fontsize=12)
plt.ylabel('产品数量', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 4. 评分分布：看产品整体口碑
# 图表类型：条形图作用：判断产品的用户评价水平
plt.figure(figsize=(10, 6))
rating_counts = data_2['rating'].value_counts().sort_index()
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='coolwarm')
plt.title('产品评分分布', fontsize=14)
plt.xlabel('评分', fontsize=12)
plt.ylabel('产品数量', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 5. 评分 vs 人气：看高口碑产品的受欢迎度
# 图表类型：散点图作用：识别 “高评分 + 高人气” 的优质产品
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data_2, 
    x='rating', 
    y='rating_count', 
    size='discount_percentage',  # 气泡大小=折扣率
    sizes=(50, 500), 
    alpha=0.7, 
    color='darkorange'
)
plt.title('评分vs评分数量（人气）', fontsize=14)
plt.xlabel('评分', fontsize=12)
plt.ylabel('评分数量（人气）', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 6. 性价比分析：选 “低价 + 高评分 + 高折扣” 产品
# 图表类型：气泡图作用：快速锁定高性价比产品（真实工作中用于选品 / 推广）
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=data_2,
    x='discounted_price',  # 折扣价（越低越优）
    y='rating',            # 评分（越高越优）
    size='rating_count',   # 气泡大小=人气
    hue='discount_percentage',  # 颜色=折扣率
    sizes=(100, 2000),
    alpha=0.6,
    palette='RdYlBu_r'
)
plt.title('产品性价比分析（折扣价+评分+人气+折扣率）', fontsize=14)
plt.xlabel('折扣价', fontsize=12)
plt.ylabel('评分', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
